```python
new_customer_count = df[df['å®¢æˆ·ç±»å‹'] == 'æ–°å®¢æˆ·'].groupby('åŸå¸‚')['è®¢å•ID'].nunique()
print("å„åŸå¸‚æ–°å®¢æˆ·æ•°é‡ï¼š")
print(new_customer_count)
```

```python
city_metrics = df.groupby('åŸå¸‚').agg({
    'é”€å”®é¢': 'sum',
    'è®¢å•ID': 'nunique'
}).rename(columns={'è®¢å•ID': 'æ€»è®¢å•æ•°'})

city_metrics['å¹³å‡å®¢å•ä»·'] = city_metrics['é”€å”®é¢'] / city_metrics['æ€»è®¢å•æ•°']
print("\nå„åŸå¸‚å¹³å‡å®¢å•ä»·ï¼š")
print(city_metrics[['å¹³å‡å®¢å•ä»·']])
```

```python
# åˆ›å»ºå¼€åº—ä¼˜å…ˆçº§æ•°æ®
priority_df = pd.DataFrame({
    'æ–°å®¢æˆ·æ•°é‡': new_customer_count,
    'å¹³å‡å®¢å•ä»·': city_metrics['å¹³å‡å®¢å•ä»·']
})

# è®¡ç®—å¼€åº—ä¼˜å…ˆçº§ = æ–°å®¢æˆ·æ•°é‡ Ã— å¹³å‡å®¢å•ä»·
priority_df['å¼€åº—ä¼˜å…ˆçº§'] = priority_df['æ–°å®¢æˆ·æ•°é‡'] * priority_df['å¹³å‡å®¢å•ä»·']

# æŒ‰ä¼˜å…ˆçº§æ’åº
priority_df = priority_df.sort_values('å¼€åº—ä¼˜å…ˆçº§', ascending=False)

print("=== å¼€åº—ä¼˜å…ˆçº§åˆ†æ ===")
print(priority_df)

print(f"\nğŸ¯ æ¨èå¼€åº—åŸå¸‚ï¼š{priority_df.index[0]}")
print(f"   - æ–°å®¢æˆ·æ•°é‡ï¼š{priority_df.iloc[0]['æ–°å®¢æˆ·æ•°é‡']}ä¸ª")
print(f"   - å¹³å‡å®¢å•ä»·ï¼šÂ¥{priority_df.iloc[0]['å¹³å‡å®¢å•ä»·']:,.0f}")
print(f"   - ä¼˜å…ˆçº§åˆ†æ•°ï¼š{priority_df.iloc[0]['å¼€åº—ä¼˜å…ˆçº§']:.0f}")
```
